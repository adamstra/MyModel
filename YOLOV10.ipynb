```python
import cv2
import locale
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Set the locale to UTF-8
def getpreferredencoding(do_setlocale = True):
    return "UTF-8"
locale.getpreferredencoding = getpreferredencoding

model = YOLO('/content/drive/MyDrive/runs/detect/Yolo_model/weights/best.pt')

# Réaliser des prédictions sur une image
results = model('/content/drive/MyDrive/Road_traffic/images.jpeg')

# Afficher les résultats
# results[0].show()


# Charger l'image originale
image_path = '/content/drive/MyDrive/Road_traffic/images.jpeg'
original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

# Obtenir l'image avec la détection
detected_image = results[0].plot()

# Afficher les deux images côte à côte
plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title('Image Originale')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(detected_image)
plt.title('Image avec Détection')
plt.axis('off')

plt.show()
```

```python
# Évaluation sur le jeu de validation
results = model.val()

# Évaluation sur le jeu de test
results = model.val(data='/content/drive/MyDrive/Road_traffic/data.yaml', split='test')

import matplotlib.pyplot as plt

# Access metrics using results_dict
metrics = results.results_dict

# Données des métriques
mAP_50 = metrics['metrics/mAP50(B)'] # Changed key to match metrics dictionary
mAP_50_95 = metrics['metrics/mAP50-95(B)'] # Changed key to match metrics dictionary
precision = metrics['metrics/precision(B)'] # Changed key to match metrics dictionary
recall = metrics['metrics/recall(B)'] # Changed key to match metrics dictionary

# Noms des métriques
metric_names = ['mAP@0.5', 'mAP@0.5:0.95', 'Précision', 'Rappel']
metric_values = [mAP_50, mAP_50_95, precision, recall]

# Afficher les métriques
plt.figure(figsize=(10, 5))
plt.bar(metric_names, metric_values, color=['blue', 'orange', 'green', 'red'])
plt.title('Performances du Modèle')
plt.ylim([0, 1])
plt.ylabel('Score')
plt.show()
```

```python

# Données des métriques
mAP_50 = metrics['metrics/mAP50(B)']
mAP_50_95 = metrics['metrics/mAP50-95(B)']
precision = metrics['metrics/precision(B)']
recall = metrics['metrics/recall(B)']

# Afficher les métriques en pourcentage
print(f"mAP@0.5: {mAP_50 * 100:.2f}%")
print(f"mAP@0.5:0.95: {mAP_50_95 * 100:.2f}%")
print(f"Précision: {precision * 100:.2f}%")
print(f"Rappel: {recall * 100:.2f}%")
```
